### Multi-Folder File Download with Progress Tracking

```python
import os
import shutil
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm
import requests
import time

# Construct download URL
def construct_download_url(file_id):
    return f"https://drive.google.com/uc?export=download&id={file_id}"

# Download a single file with progress tracking
def download_file(file_path, download_url, session, progress_bar):
    try:
        response = session.get(download_url, stream=True)
        total_size = int(response.headers.get('content-length', 0))
        block_size = 8192  # Define block size
        with open(file_path, 'wb') as file:
            for data in response.iter_content(block_size):
                file.write(data)
                progress_bar.update(len(data))
    except Exception as e:
        print(f"Error downloading {file_path}: {e}")

# Process a single folder and download its files
def process_folder(folder, session, progress_bar):
    folder_name = folder['title']
    folder_id = folder['id']
    print(f"Processing folder: {folder_name}")

    # Create folder directory
    folder_download_path = os.path.join(download_destination, folder_name)
    os.makedirs(folder_download_path, exist_ok=True)

    # Fetch file metadata
    files = drive.ListFile({'q': f"'{folder_id}' in parents and trashed=false"}).GetList()
    download_tasks = []
    for file_info in files:
        if file_info['mimeType'] != 'application/vnd.google-apps.folder':  # Skip subfolders
            file_name = file_info['title']
            file_path = os.path.join(folder_download_path, file_name)
            download_url = construct_download_url(file_info['id'])
            download_tasks.append((file_path, download_url, session, progress_bar))

    # Execute downloads in parallel
    with ThreadPoolExecutor(max_workers=os.cpu_count() * 2) as executor:
        executor.map(lambda args: download_file(*args), download_tasks)

# Main execution
if __name__ == "__main__":
    session = requests.Session()  # Shared session for all requests

    # Reset download destination
    download_destination = "/content/Augmented_data/preprocessed_data"
    if os.path.exists(download_destination):
        shutil.rmtree(download_destination)
    os.makedirs(download_destination)

    # Calculate total size for progress bar
    total_size = 0
    for folder in file_list:
        files = drive.ListFile({'q': f"'{folder['id']}' in parents and trashed=false"}).GetList()
        for file_info in files:
            if file_info['mimeType'] != 'application/vnd.google-apps.folder':  # Only count files
                total_size += int(file_info.get('fileSize', 0))  # Use `fileSize` from metadata

    # Measure start time
    start_time = time.time()

    # Single progress bar for all downloads
    with tqdm(total=total_size, unit='iB', unit_scale=True, desc="Downloading All Files") as progress_bar:
        for folder in file_list:
            process_folder(folder, session, progress_bar)

    # Measure end time
    end_time = time.time()
    total_time = end_time - start_time

    # Display execution time
    print(f"Download completed in {total_time:.2f} seconds.")
```

# Mirror flipping

```python
import os
import cv2
import shutil
from tqdm import tqdm

def mirror_video(video_path, output_path):
    try:
        # Open the video file
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"Error opening video file: {video_path}")
            return

        # Get video properties
        fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # MP4 format

        # Ensure output folder exists
        os.makedirs(os.path.dirname(output_path), exist_ok=True)

        # Initialize video writer
        out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        # Process frames and mirror
        with tqdm(total=total_frames, desc=f"Mirroring {os.path.basename(video_path)}", unit='frame', ncols=100) as pbar:
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                # Mirror the frame horizontally
                mirrored_frame = cv2.flip(frame, 1)
                out.write(mirrored_frame)
                pbar.update(1)

        # Release resources
        cap.release()
        out.release()
        print(f"Mirrored video saved: {output_path}")

    except Exception as e:
        print(f"Error processing video {video_path}: {e}")

def mirror_all_videos(input_folder, output_folder):
    # Collect all video files
    video_files = []
    for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.lower().endswith(('mp4', 'avi', 'mov')):  # Supported formats
                video_files.append(os.path.join(root, file))

    with tqdm(total=len(video_files), desc="Processing Videos", unit='video', ncols=100) as video_pbar:
        for video_path in video_files:
            # Generate mirrored output path
            relative_path = os.path.relpath(video_path, input_folder)
            output_path = os.path.join(output_folder, relative_path)
            mirror_video(video_path, output_path)
            video_pbar.update(1)

if __name__ == "__main__":
    # Define input and output paths
    root = "/content/Augmented_data/preprocessed_data"
    destination = "/content/Augmented_data/mirrored"

    # Ensure destination folder is clean
    if os.path.exists(destination):
        print(f"Removing existing destination folder: {destination}")
        shutil.rmtree(destination)
    os.makedirs(destination, exist_ok=True)

    # Start mirroring videos
    print(f"Mirroring videos from {root} to {destination}")
    mirror_all_videos(root, destination)
```

## Rotate by 90 degree

```python
import os
import cv2
import shutil
from tqdm import tqdm


def rotate_video_90(video_path, output_folder):
    try:
        print(f"Processing video: {video_path}")

        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"Error opening video file: {video_path}")
            return

        fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        # Define codec and output video path
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        class_folder = os.path.basename(os.path.dirname(video_path))
        class_output_folder = os.path.join(output_folder, class_folder)
        os.makedirs(class_output_folder, exist_ok=True)
        output_path = os.path.join(class_output_folder, os.path.basename(video_path))

        # Set up VideoWriter with swapped dimensions
        out = cv2.VideoWriter(output_path, fourcc, fps, (height, width))

        # Process video frames
        with tqdm(total=total_frames, desc=f"Rotating {os.path.basename(video_path)}", unit='frame', ncols=100) as pbar:
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                # Rotate frame 90 degrees clockwise
                rotated_frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
                out.write(rotated_frame)
                pbar.update(1)

        cap.release()
        out.release()
        print(f"Video rotated and saved: {output_path}")

    except Exception as e:
        print(f"Error rotating video {video_path}: {e}")


def rotate_videos_90(root, output_folder):
    video_files = []
    for root_dir, dirs, files in os.walk(root):
        for file in files:
            if file.lower().endswith(('mp4', 'avi', 'mov')):  # Filter video files
                video_files.append(os.path.join(root_dir, file))

    with tqdm(total=len(video_files), desc="Rotating Videos by 90°", unit='video', ncols=100) as video_pbar:
        for video_path in video_files:
            rotate_video_90(video_path, output_folder)
            video_pbar.update(1)


if __name__ == "__main__":
    root = "/content/Augmented_data/preprocessed_data"
    output_folder = "/content/Augmented_data/rotated_90"

    # Ensure the output folder is clean
    if os.path.exists(output_folder):
        print(f"Removing existing rotated_90 folder: {output_folder}")
        shutil.rmtree(output_folder)
    os.makedirs(output_folder)

    print(f"Starting rotation for videos in {root}")
    rotate_videos_90(root, output_folder)
```

## Brightness adjustment (increasing the brightness)

```python
import os
import cv2
import shutil
from tqdm import tqdm

def adjust_brightness_video(video_path, output_path, brightness_factor):
    """
    Adjust the brightness of a video and save it to the output path.

    Args:
    - video_path: Path to the input video.
    - output_path: Path to save the brightness-adjusted video.
    - brightness_factor: Factor to adjust brightness (e.g., >1 for brighter, <1 for darker).
    """
    try:
        # Open the video file
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"Error opening video file: {video_path}")
            return

        # Get video properties
        fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # MP4 format

        # Ensure output folder exists
        os.makedirs(os.path.dirname(output_path), exist_ok=True)

        # Initialize video writer
        out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        # Process frames and adjust brightness
        with tqdm(total=total_frames, desc=f"Processing {os.path.basename(video_path)}", unit='frame', ncols=100) as pbar:
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                # Adjust brightness
                adjusted_frame = cv2.convertScaleAbs(frame, alpha=brightness_factor, beta=0)
                out.write(adjusted_frame)
                pbar.update(1)

        # Release resources
        cap.release()
        out.release()
        print(f"Brightness-adjusted video saved: {output_path}")

    except Exception as e:
        print(f"Error processing video {video_path}: {e}")

def adjust_brightness_all_videos(input_folder, output_folder, brightness_factor):
    """
    Adjust the brightness of all videos in the input folder and save them in the output folder.
    """
    video_files = []
    for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.lower().endswith(('mp4', 'avi', 'mov')):  # Supported formats
                video_files.append(os.path.join(root, file))

    with tqdm(total=len(video_files), desc="Adjusting Brightness", unit='video', ncols=100) as video_pbar:
        for video_path in video_files:
            # Generate output path
            relative_path = os.path.relpath(video_path, input_folder)
            output_path = os.path.join(output_folder, relative_path)
            adjust_brightness_video(video_path, output_path, brightness_factor)
            video_pbar.update(1)

if __name__ == "__main__":
    # Define input and output paths
    root = "/content/Augmented_data/preprocessed_data"  # Input folder containing videos
    output = "/content/Augmented_data/increased_brightness"  # Output folder for adjusted videos
    brightness_factor = 1.5  # >1 for brighter, <1 for darker (e.g., 1.5 for 50% brighter)

    # Ensure the output folder is clean
    if os.path.exists(output):
        print(f"Removing existing folder: {output}")
        shutil.rmtree(output)
    os.makedirs(output, exist_ok=True)

    # Adjust brightness
    print(f"Adjusting brightness of videos in {root}")
    adjust_brightness_all_videos(root, output, brightness_factor)
```

## Vignette

```python
import os
import cv2
import numpy as np
from tqdm import tqdm
import shutil

def apply_vignette(frame, intensity=3):
    """
    Apply a vignette effect to a single frame with adjustable intensity.

    Args:
    - frame: Input frame from the video.
    - intensity: Controls the intensity of the vignette effect (higher = darker edges).
    """
    rows, cols = frame.shape[:2]

    # Create Gaussian kernels for x and y directions
    x = cv2.getGaussianKernel(cols, cols / intensity)  # Smaller sigma for sharper edges
    y = cv2.getGaussianKernel(rows, rows / intensity)
    kernel = y * x.T
    mask = kernel / kernel.max()

    # Apply the mask to each channel
    vignette_frame = np.zeros_like(frame)
    for i in range(3):  # For BGR channels
        vignette_frame[:, :, i] = frame[:, :, i] * mask

    return vignette_frame.astype(np.uint8)


def process_video(video_path, output_folder, intensity=3):
    """
    Process a video by applying vignette effect to each frame and save it.

    Args:
    - video_path: Path to the input video.
    - output_folder: Folder to save the processed video.
    - intensity: Vignette intensity.
    """
    try:
        print(f"Processing video: {video_path}")

        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"Error opening video file: {video_path}")
            return

        fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        # Define codec and output video path
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        class_folder = os.path.basename(os.path.dirname(video_path))
        class_output_folder = os.path.join(output_folder, class_folder)
        os.makedirs(class_output_folder, exist_ok=True)
        output_path = os.path.join(class_output_folder, os.path.basename(video_path))

        # Set up VideoWriter
        out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        # Process video frames and apply vignette
        with tqdm(total=total_frames, desc=f"Applying Vignette to {os.path.basename(video_path)}", unit='frame', ncols=100) as pbar:
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                # Apply vignette effect to the frame
                vignette_frame = apply_vignette(frame, intensity)

                # Write the frame to the output video
                out.write(vignette_frame)
                pbar.update(1)

        cap.release()
        out.release()
        print(f"Vignette applied video saved: {output_path}")

    except Exception as e:
        print(f"Error processing video {video_path}: {e}")


def process_all_videos(root, output_folder, intensity=3):
    """
    Apply vignette effect to all videos in the root folder and save them to the output folder.

    Args:
    - root: Root folder containing videos to be processed.
    - output_folder: Output folder to save the processed videos.
    - intensity: Vignette intensity.
    """
    video_files = []
    for root_dir, dirs, files in os.walk(root):
        for file in files:
            if file.lower().endswith(('mp4', 'avi', 'mov')):  # Filter video files
                video_files.append(os.path.join(root_dir, file))

    with tqdm(total=len(video_files), desc="Processing Videos", unit='video', ncols=100) as video_pbar:
        for video_path in video_files:
            # Process each video
            process_video(video_path, output_folder, intensity)
            video_pbar.update(1)


if __name__ == "__main__":
    # Define input and output paths
    root = "/content/Augmented_data/preprocessed_data"  # Input folder containing the original videos
    output_folder = "/content/Augmented_data/vignette"  # Folder to save processed videos

    # Ensure the output folder is clean
    if os.path.exists(output_folder):
        print(f"Removing existing folder: {output_folder}")
        shutil.rmtree(output_folder)
    os.makedirs(output_folder, exist_ok=True)

    # Apply vignette effect to all videos in the input folder
    print(f"Applying vignette effect to videos in {root}")
    process_all_videos(root, output_folder, intensity=3)
```

## Blurred

```python
import os
import cv2
import numpy as np
from tqdm import tqdm
import shutil

def apply_blur(frame, blur_strength=15):
    """
    Apply blur effect to a single frame with specified blur strength.

    Args:
    - frame: Input frame from the video.
    - blur_strength: The strength of the blur (larger values = stronger blur).
    """
    # Apply GaussianBlur to the frame with the specified kernel size
    return cv2.GaussianBlur(frame, (blur_strength, blur_strength), 0)


def process_video(video_path, output_folder, blur_strength=15):
    """
    Process a video by applying blur effect to each frame and save it.

    Args:
    - video_path: Path to the input video.
    - output_folder: Folder to save the processed video.
    - blur_strength: Strength of the blur effect.
    """
    try:
        print(f"Processing video: {video_path}")

        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"Error opening video file: {video_path}")
            return

        fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        # Define codec and output video path
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        class_folder = os.path.basename(os.path.dirname(video_path))
        class_output_folder = os.path.join(output_folder, class_folder)
        os.makedirs(class_output_folder, exist_ok=True)
        output_path = os.path.join(class_output_folder, os.path.basename(video_path))

        # Set up VideoWriter
        out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        # Process video frames and apply blur
        with tqdm(total=total_frames, desc=f"Applying Blur to {os.path.basename(video_path)}", unit='frame', ncols=100) as pbar:
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                # Apply blur effect to the frame
                blurred_frame = apply_blur(frame, blur_strength)

                # Write the frame to the output video
                out.write(blurred_frame)
                pbar.update(1)

        cap.release()
        out.release()
        print(f"Blurred video saved: {output_path}")

    except Exception as e:
        print(f"Error processing video {video_path}: {e}")


def process_all_videos(root, output_folder, blur_strength=15):
    """
    Apply blur effect to all videos in the root folder and save them to the output folder.

    Args:
    - root: Root folder containing videos to be processed.
    - output_folder: Output folder to save the processed videos.
    - blur_strength: Strength of the blur effect.
    """
    video_files = []
    for root_dir, dirs, files in os.walk(root):
        for file in files:
            if file.lower().endswith(('mp4', 'avi', 'mov')):  # Filter video files
                video_files.append(os.path.join(root_dir, file))

    with tqdm(total=len(video_files), desc="Processing Videos", unit='video', ncols=100) as video_pbar:
        for video_path in video_files:
            # Process each video
            process_video(video_path, output_folder, blur_strength)
            video_pbar.update(1)


if __name__ == "__main__":
    # Define input and output paths
    root = "/content/Augmented_data/preprocessed_data"  # Input folder containing the original videos
    output_folder = "/content/Augmented_data/blurred"  # Folder to save processed videos

    # Ensure the output folder is clean
    if os.path.exists(output_folder):
        print(f"Removing existing folder: {output_folder}")
        shutil.rmtree(output_folder)
    os.makedirs(output_folder, exist_ok=True)

    # Apply blur effect to all videos in the input folder
    print(f"Applying blur effect to videos in {root}")
    process_all_videos(root, output_folder, blur_strength=15)
```
